```
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler,OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```
data = load_boston()

data
```

```
inputdata = data.data
targetdata = data.target
```

```
scaler = StandardScaler()
inputscaled = scaler.fit_transform(inputdata)
x_train,x_test,y_train,y_test = train_test_split(inputdata,targetdata,test_size=0.3)
```

```
model = Sequential()
h1 = model.add(Dense(20,activation="relu",input_dim=13))
h2=model.add(Dense(20,activation="relu"))
h3=model.add(Dense(20,activation="relu"))
o = model.add(Dense(1,activation="linear"))
```

```
model.compile(loss="mean_squared_error",optimizer="Adam",metrics=['accuracy'])
```

```
model.summary()
```

```
model.fit(x_train,y_train,batch_size=25,epochs=10000
```

```
score = model.evaluate(x_test,y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
